Import the modules necessary for simulation

```python
from photonmap.Simulator import *

from pgljupyter import SceneWidget
from openalea.plantgl.all import *
import random

spheres = []

for i in range(255):
    x, y, z = [random.choice([-1, 1]) * random.randrange(0, int(255/2)) for i in range(3)]
    m = Material(ambient=(x+int(255/2), y+int(255/2), z+int(255/2)), shininess=0.1, diffuse=1)
    spheres.append(Shape(Translated(x, y, z, Sphere(random.randrange(1, 20), 20, 20)), m))
    
SceneWidget(spheres, size_display=(600, 400), size_world=255)
```

Setup configuration

```python
simulator = Simulator()
simulator.readConfiguration("simulation.ini")

simulator.setupRoom("./assets/testChamber.rad", "./PO")
simulator.setupCaptor("./captors/captors_expe1.csv")
#simulator.setupPlant("./assets/rose-simple4.lpy", Vec3(1280.0, 860.0, 980.0))
```

Run the simulation

```python

res = simulator.run()
calibrated_res = simulator.calibrateResults("spectrum/chambre1_spectrum", "points_calibration.csv")

simulator.writeResults()
```

Test Tmin in complete scene

```python
simulator.test_t_min(int(1e6), 1e-6, 11, False)
```
